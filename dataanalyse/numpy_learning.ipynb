```python
import numpy as np

np.__version__


def numpy_sum(n):
    a = np.arange(n) ** 2
    b = np.arange(n) ** 3
    return a + b


def python_sum(n):
    a = [i ** 2 for i in range(n)]
    b = [i ** 3 for i in range(n)]
    ab_sum = []
    for i in range(n):
        ab_sum.append(a[i] + b[i])


print(numpy_sum(10))

%timeit numpy_sum(1000)
%timeit python_sum(1000)
```

```python
%timeit numpy_sum(100000)
%timeit python_sum(100000)
```

```python
%timeit numpy_sum(1000000)
%timeit python_sum(1000000)
```

### Numpy ndarray

ndarray是一个n维数组对象

```python
arr1 = np.array([1,2,3,4,5])
print(type(arr1))

arr2 = np.array((1, 2, 3, 4, 5))
print(type(arr2))

arr3 = np.array(range(10))
print(arr3)

arr4 = np.array([i**2 for i in range(10)])
print(arr4)

arr5 = np.array([i for i in range(10) if i % 2 == 0])
print(arr5)
```

```python
arr6 = np.arange(0, 10, 2)
print(arr6)

print(np.arange(5, dtype=float))


print(np.linspace(0, 4, 9))
print(np.linspace(0, 4.1, 9))

print(np.linspace(2.0, 3.0, 5))

arr7 = np.linspace(2.0, 3.0, num=5, retstep=True)
print(arr7)
print(type(arr7))

print(np.linspace(0.0, 1.0, num=10, retstep=True))

```

```python
a = np.arange(20).reshape((4,5))
print(a)

print(a.itemsize)
```

```python
arr1 = np.arange(10)
print(arr1)

arr2 = arr1[2:7:2]
print(arr2)

a = np.arange(20).reshape((4,5))

print(a[...,1])

print(a[[0, 1, 2], [0, 1, 0]])


z = np.zeros((8, 8), dtype=int)
print(z)

z[1::2, ::2] = 1
z[::2,1::2] = 1
print(z)
```

```python
b = np.arange(12).reshape((3,4))

print(b[b > 6])
print(b[b % 2 == 1])

b[b % 2 == 1] = -1
print(b)

print(b[[False, True, True],[True, False, True, False]])
print('----------------------------------')
print(b)

print((b[[1,-1], ...]))
```

```python
a = np.array([1,2,3,4])
b = np.array([10, 20, 30, 40])
print(a*b)

c = np.array([[1,2,3],[4,5,6]])
d = np.array([1,2,3])

print(c+d)

arr = np.arange(20).reshape((4,5))
print(arr)
arr.mean(axis=0)

```

```python
a = np.array([95, 85, 75, 75, 55, 45])
b = np.array([73, 72, 71, 69, 68, 67])

print(np.std(a))
print(np.std(b))

print(np.var(a))
print(np.var(b))

xiaoming = np.array([80,90, 95])
xiaogang = np.array([95,90, 80])

weight = np.array([0.2, 0.3, 0.5])

print(np.average(xiaogang, weights=weight))
print(np.average(xiaoming, weights=weight))
```
